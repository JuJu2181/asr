Updating the columns to check presence of a character in label. If 1 means it means present if 0 it means absent

```python
def checkPresenceAndTotal(df):
    print(f'There are total {df.shape[0]} rows in the dataset')
    print('Iteration started')
    for i in range(df.shape[0]):
        for c in df.iloc[i]['text']:
            if c == ' ' or c == '\u200c' or c == '\u200d': c = 'space'
            if c not in unique_chars: continue
            df.iloc[i,df.columns.get_loc(c)] = 1
        if i % 10000 == 0:
            print(f'{i} rows done')
    print('Iteration Completed.\nNow Calculating total')
    df['total'] = df.iloc[:,2:].sum(axis=1) 
    print('Total Calculated\nProcess Completed!!')
    return df 
```
